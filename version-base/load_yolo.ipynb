```python
# Import Libraries
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import xml.etree.ElementTree as ET
import seaborn as sns
from tqdm import tqdm
from ultralytics import YOLO
from datetime import datetime
import matplotlib as mpl
```

```python
YOLO_BASE_VERSION = 'yolov8s.yaml'
WEIGHTS_PATH = '../runs/detect/gcolab_china_motorbike_20ex16b/weights/best.pt'

# Load a pretrained YOLO model (recommended for training)
model = YOLO(WEIGHTS_PATH)
```

```python
# SINGLE IMAGE PREDICTION
# Predict with the trained model
test_image = {
    1: '../datasets/RDD2022/images/China_MotorBike/test/images/China_MotorBike_001987.jpg'
}

train_images = {
    1: '../datasets/RDD2022/images/China_MotorBike/train/images/China_MotorBike_001356.jpg'
}

selected_image = train_images[1]

# Load the image
img = cv2.imread(selected_image)

# Predict and show image with predicted bounding boxes
results = model.predict(source=img, save=True)
```
