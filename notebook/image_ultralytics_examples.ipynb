# Gráficas de Ultralytics

En este notebook vamos a juntar varios ejemplos de gráficas que genera ultralytics durante el entrenamiento para tener una única imagen que se pueda añadir a la memoria del TFG.

## Importar librerías

```python
import os, sys
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

## Configuración

```python
PATH = '../memoria/graphs/'

IMAGES = [
    'metrics_confusion_matrix_normalized.png',
    'metrics_PR_curve.png',
    'metrics_results.png',
    'metrics_val_batch0_pred.jpg',
]
```

## Creamos la gráfica

```python
# Creamos una sola gráfica con las cuatro imágenes en una grid de 2x2
fig, axs = plt.subplots(2, 2, figsize=(20, 15))

for i, img in enumerate(IMAGES):
    img = cv2.imread(PATH + img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axs[i//2, i%2].imshow(img)
    axs[i//2, i%2].axis('off')

plt.tight_layout()
# Guardamos la imagen
plt.savefig(PATH + 'metrics_grid.jpg')
plt.show()
```
